```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml import Pipeline

# Create synthetic distributed dataset
data = [
    Row(age=25, salary=50000, experience=2, label=0),
    Row(age=30, salary=60000, experience=5, label=0),
    Row(age=45, salary=80000, experience=20, label=1),
    Row(age=50, salary=90000, experience=25, label=1),
    Row(age=23, salary=48000, experience=1, label=0),
    Row(age=40, salary=75000, experience=15, label=1),
]

df = spark.createDataFrame(data)

# Feature Engineering
assembler = VectorAssembler(
    inputCols=["age", "salary", "experience"],
    outputCol="features"
)

# Model
lr = LogisticRegression(featuresCol="features", labelCol="label")

# Pipeline
pipeline = Pipeline(stages=[assembler, lr])

# Train/Test split
train, test = df.randomSplit([0.8, 0.2], seed=42)

# Train
model = pipeline.fit(train)

# Predict
predictions = model.transform(test)

predictions.select("age", "salary", "experience", "prediction", "probability").show()

# Evaluate
evaluator = BinaryClassificationEvaluator(labelCol="label")
auc = evaluator.evaluate(predictions)

print("AUC:", auc)
```

```python
from pyspark.sql import Row
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression

# Sample dataset
data = [
    Row(age=25, salary=50000, experience=2, label=0),
    Row(age=30, salary=60000, experience=5, label=0),
    Row(age=35, salary=65000, experience=7, label=1),
    Row(age=45, salary=80000, experience=20, label=1),
    Row(age=40, salary=75000, experience=15, label=1),
]

df = spark.createDataFrame(data)

assembler = VectorAssembler(
    inputCols=["age", "salary", "experience"],
    outputCol="features"
)

df2 = assembler.transform(df)

lr = LogisticRegression(featuresCol="features", labelCol="label")

model = lr.fit(df2)

print("Coefficients:", model.coefficients)
print("Intercept:", model.intercept)
```

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml import Pipeline
from pyspark.ml.classification import LogisticRegression

assembler = VectorAssembler(
    inputCols=["age", "salary", "experience"],
    outputCol="raw_features"
)

scaler = StandardScaler(
    inputCol="raw_features",
    outputCol="features",
    withStd=True,
    withMean=True
)

lr = LogisticRegression(labelCol="label")

pipeline = Pipeline(stages=[assembler, scaler, lr])

model = pipeline.fit(df)

lr_model = model.stages[-1]

print("Scaled Coefficients:", lr_model.coefficients)
print("Intercept:", lr_model.intercept)
```

```python
from pyspark.sql.functions import col
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline

# Create base DataFrame once
base_df = spark.range(0, 1000000)

# Derive columns from SAME dataframe
large_df = base_df \
    .withColumn("age", (col("id") % 50) + 20) \
    .withColumn("salary", (col("id") % 100) * 1000 + 30000) \
    .withColumn("experience", (col("id") % 30)) \
    .withColumn("label", (col("id") % 2).cast("double")) \
    .drop("id")

assembler = VectorAssembler(
    inputCols=["age", "salary", "experience"],
    outputCol="features"
)

lr = LogisticRegression(maxIter=10)

pipeline = Pipeline(stages=[assembler, lr])

model = pipeline.fit(large_df)

print("Model trained on 1,000,000 rows successfully.")
```
